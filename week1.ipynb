### 1. Write a Python program to construct the following pattern, using a nested for loop. <br>
Input: <br>
n = 3 <br>
<br>
Output: <br>
x <br>
xx <br>

n = 10

```python
n = 10
for x in range(n):
  for y in range(x):
    print ("*",end="")
  print()
```
